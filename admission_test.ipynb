```python
col_dtypes = df.dtypes.tolist()
numeric_cols = [df.columns[idx] for idx, col_type in enumerate(col_dtypes) if col_type=='int64' or col_type=='float64']
```

```python
df_numerics = df[numeric_cols]
df_numerics.head()
```

```python
#Simple feature scaling
df_scale1 = df_numerics/df_numerics.max()
df_scale1.head()
```

```python
#Z-Score
df_scale2 = (df_numerics-df_numerics.mean())/(df_numerics.std())
df_scale2.head(10)
```

```python
df_scale2.mean()
```

```python
df_scale2.std()
```
